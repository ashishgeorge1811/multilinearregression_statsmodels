```python
from sklearn.model_selection import train_test_split
cereal_train, cereal_test  = train_test_split(cereal,test_size = 0.2)
```

```python
model_train = smf.ols("rating~calories+protein+fat+sodium+fiber+carbo+sugars+potass+vitamins",data=cereal_train).fit()
model_train.summary()
```

```python
test_predict = model_train.predict(cereal_test)
test_predict.head()
```

```python
cereal_test.rating.head()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse_test = sqrt(mean_squared_error(cereal_test.rating, test_predict))
rmse_test
```

```python
train_predict = model_train.predict(cereal_train)
train_predict.head()
```

```python
cereal_train.rating.head()
```

```python
rmse_train = sqrt(mean_squared_error(cereal_train.rating, train_predict))
rmse_train
```
